```python
class bank:
    def __init__(self):
        self.name = ""
        self.bal = 0
        self.pin = ""
        self.menu()

    def menu(self):
        option = input('''
        1. Press 1 for setting the Balance
        2. Press 2 for pin
        3. Press 3 for Name
        4. Press 4 for withdrawl
        ''')

        if option == "1":
            self.set_balance()
        elif option == "2":
            self.set_pin()
        elif option == "3":
            self.enter_name()
        elif option == "4":
            self.withdrawl()

    def set_balance(self):
        new_bal = float(input("Enter your balance: "))
        self.bal = new_bal
        print("Balance Updated!")
        self.menu()

    def set_pin(self):
        pin = int(input("Enter Your pin: "))
        self.pin = pin
        print("PIN Updated!")
        self.menu()

    def enter_name(self):
        name = input("Enter Your name: ")
        self.name = name
        print("Name Updated!")
        self.menu()

    def withdrawl(self):
        attempts = 0  # Wrong PIN counter

        while attempts < 3:
            pin = int(input("Enter your pin: "))

            if self.pin == pin:
                amount = int(input("Enter the Amount: "))

                if self.bal >= amount:
                    self.bal -= amount
                    print("Transaction Successful!")
                    print(f"Remaining Amount: {self.bal}")
                else:
                    print("Insufficient Balance!")
                break   # Correct pin found → exit loop

            else:
                attempts += 1
                print("Wrong PIN!")

                if attempts == 3:
                    print("⚠️ Your account has been temporarily blocked due to 3 incorrect attempts.")
                    return    # ← Yaha koi syntax error nahi hai
                else:
                    print(f"Try again... Attempts left: {3 - attempts}")

        self.menu()
```
